```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error

def XGBoost_Model(X_train, X_test, y_train, y_test):
    xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
    xgb.fit(X_train, y_train)

    train_predict = xgb.predict(X_train)
    test_predict = xgb.predict(X_test)

    train_mse = mean_squared_error(y_train, train_predict)
    test_mse = mean_squared_error(y_test, test_predict)
    train_r2 = r2_score(y_train, train_predict)
    test_r2 = r2_score(y_test, test_predict)

    print("\nEvaluation with XGBoost Regressor")
    print(f"Train Error = {train_mse*100} Test Error = {test_mse*100}")
    print(f"Train Score = {train_r2*100} Test Score = {test_r2*100}")

    return xgb

# Assuming X_train, X_test, y_train, and y_test are defined
model_xgb = XGBoost_Model(X_train, X_test, y_train, y_test)
```

```python
from scipy.stats import norm
import numpy as np

def zscore_to_percentage(z_scores):
    probabilities = norm.cdf(z_scores)
    percentages = probabilities * 100
    return percentages

def print_zscore_percentages(y_predicted):
    percentages = zscore_to_percentage(y_predicted)

    for z_score, percentage in zip(y_predicted, percentages):
        print(f"A Z-score of {z_score:.2f} corresponds to a percentage of {percentage:.2f}%")

y_predicted = np.array([1.0860448, 1.9056444, -0.5385434, 0.02713531, -0.33408263,
                        -1.413009, 0.97896075, -0.605784, -1.5082855, 0.46538755], dtype=np.float32)

print_zscore_percentages(y_predicted)
```
